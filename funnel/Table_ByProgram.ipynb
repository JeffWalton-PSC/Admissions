```python
df = pd.read_hdf("X:/Applications/Admissions/funnel/data/stage_data", key="weekly")
df = df.loc[df["year_term"].isin([this_yearterm, last_yearterm, twoyrago_yearterm])]
```

```python
def adm_week(d):
    """
    returns calendar week number and Admissions Week Number for a given date, d
    """
    year = d.year
    if d >= date(year, 9, 1):
        adm_year_start = year
    else:
        adm_year_start = year - 1

    week_number = d.isocalendar()[1]
    adm_start_week = date(adm_year_start, 9, 1).isocalendar()[1]

    if week_number >= adm_start_week:
        adm_week_number = week_number - adm_start_week
    else:
        adm_week_number = 53 + (week_number - adm_start_week)

    return (week_number, adm_week_number)
```

```python
week_number, adm_week_number = adm_week(today)
week = adm_week(today)[1]
```

```python
stage = "Deposited"
plot_df = pd.DataFrame({this_yearterm: summ.loc[(this_yearterm, stage)][str(week)]})
print(plot_df.shape)
print(plot_df.head())
print(plot_df.info())
print(pd.DataFrame({last_yearterm: summ.loc[(last_yearterm, stage)][str(week)]}).info())
plot_df = (
    pd.concat(
        [
            plot_df,
            pd.DataFrame({last_yearterm: summ.loc[(last_yearterm, stage)][str(week)]}),
        ],
        axis=1,
        sort=True,
    )
    .fillna(0)
    .astype(int)
)
print(plot_df.shape)
print(plot_df)
plot_df.head()
```

```python
tab_df = pd.DataFrame(
    {this_yearterm + "_Week" + str(week): summ.loc[(this_yearterm,)][str(week)]}
)
tab_df = pd.concat(
    [
        tab_df,
        pd.DataFrame(
            {last_yearterm + "_Week" + str(week): summ.loc[(last_yearterm,)][str(week)]}
        ),
    ],
    axis=1,
    sort=True,
)
tab_df = pd.concat(
    [
        tab_df,
        pd.DataFrame(
            {
                twoyrago_yearterm
                + "_Week"
                + str(week): summ.loc[(twoyrago_yearterm,)][str(week)]
            }
        ),
    ],
    axis=1,
    sort=True,
)
tab_df = pd.concat(
    [
        tab_df,
        pd.DataFrame({last_yearterm + "_Total": summ.loc[(last_yearterm,)]["53"]}),
    ],
    axis=1,
    sort=True,
)
tab_df = pd.concat(
    [
        tab_df,
        pd.DataFrame(
            {twoyrago_yearterm + "_Total": summ.loc[(twoyrago_yearterm,)]["53"]}
        ),
    ],
    axis=1,
    sort=True,
)
tab_df = tab_df.fillna(0).astype(int)
print(tab_df.shape)
print(tab_df)
tab_df.head()
```
