## Target X data

```python
tgtX_path = Path(r"C:\JW\Python\Admissions\funnel\data\TargetX")
tgtX_file = tgtX_path / "20201222Admit.xlsx"

tgtX = pd.read_excel(tgtX_file)

print(tgtX.shape)
```

```python
tgtX.head()
```

```python
print(tgtX.columns)
```

```python
keep_cols = [
    'PowerCampus ID',
    'Anticipated Start Term Year',
    'Entered in PowerCampus',
    'First Name',
    'Last Name',
    'Student ID',
    'Anticipated Major',
    'Anticipated Start Term',
    'Anticipated Start Year',
    'Inquiry Date',
    'Applicant Date',
    'Accepted Date',
    'Deposited Date',
    'Deferred Date',
    'Withdrawn Date',
]

tgtX = tgtX.loc[:,keep_cols]

print(tgtX.shape)
tgtX.head()
```

```python
tgtX['year_term'] = (tgtX['Anticipated Start Year'].astype(str) + '.' + tgtX['Anticipated Start Term'].str.title())
tgtX['PEOPLE_CODE_ID'] = "P" + tgtX['PowerCampus ID']
```

```python
print(tgtX.shape)
tgtX.head()
```

```python
tgtX_accepts = tgtX[['PEOPLE_CODE_ID', 'year_term']]

print(tgtX_accepts.shape)

tgtX_accepts.head()
```

```python
tgtX_accepts.groupby(['year_term']).count()
```

## Differences

```python
big = pc_accepts[['PEOPLE_CODE_ID', 'year_term']].merge(tgtX_accepts, how='outer', on=['PEOPLE_CODE_ID', 'year_term'], indicator=True )

print(big.shape)
big.head()
```

```python
print(today)
writer = pd.ExcelWriter(
    f"PowerCampus-TargetX_Compare_{today}.xlsx"
)
```

```python
big.to_excel(writer, "all")

only_pc = big[(big['_merge']=='left_only')]
only_pc.to_excel(writer, "only_pc")

only_tgtX = big[(big['_merge']=='right_only')]
only_tgtX.to_excel(writer, "only_tgtX")

both = big[(big['_merge']=='both')]
both.to_excel(writer, "both")



writer.save()
```
